```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def analog_signal(f1, f2, f3):
    w1 = 2 * np.pi / f1
    w2 = 2 * np.pi / f2
    w3 = 2 * np.pi / f3

    return lambda t: np.sin(w1 * t) + np.sin(w2 * t) + np.sin(w3 * t)

def get_sample(signal_to_sample, sample_rate):
    m = int(len(signal_to_sample) / (10 * sample_rate))
    return signal_to_sample[::m]
```

```python
x = np.linspace(0, 10, 10000)
signal = analog_signal(1, 2, 3)(x)

signal_sample = get_sample(signal, 8)
x_sample = np.linspace(0,10, len(signal_sample))

plt.plot(x, signal)
plt.hlines(np.exp(-3/20) * np.max(signal), 0, 10, color="k")
plt.plot(x_sample, signal_sample, 'or')
```
